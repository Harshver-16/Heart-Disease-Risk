```python
# Rename the column '98.6' to 'oxygen' for better readability and usability
data2.rename(columns={'98.6': 'oxygen'}, inplace=True)

# Set the theme for seaborn plots to 'darkgrid' for better visual clarity
sns.set_theme(style="darkgrid")

# Set the size of the figure to 6x4 inches
plt.figure(figsize=(6, 4))

# Create a histogram with 20 bins and a KDE curve for the 'oxygen' column
sns.histplot(data2['oxygen'], bins=20, kde=True, color='dodgerblue')

# Set the title and axis labels of the plot
plt.title("Oxygen Saturation - Histogram")
plt.xlabel("Oxygen (%)")
plt.ylabel("Count")

# Display the plot
plt.show()
```

Oxygen saturation (SpO₂) is the percentage of oxygen bound to hemoglobin in the blood. It reflects how well oxygen is being delivered to the body’s organs and tissues. Normal SpO₂ levels typically range from 95% to 100%. Levels below 90% indicate hypoxemia, a condition that may require medical attention. SpO₂ is commonly measured using a pulse oximeter, a non-invasive device placed on a finger or earlobe.

```python

# Create a count plot of the 'sex' column using pastel colors
ax = sns.countplot(x='sex', data=data, palette='pastel')

# Replace default tick labels (0 and 1) with 'Female' and 'Male'
plt.xticks([0, 1], ['Female', 'Male'])

# Set the title and axis labels
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# Annotate each bar with its count value
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',              # Text: the height (count) of the bar
                (p.get_x() + p.get_width() / 2,        # X-coordinate: center of the bar
                 p.get_height() + 1),                  # Y-coordinate: slightly above the top of the bar
                ha='center', va='bottom')              # Center align the text horizontally and align it to the bottom vertically

# Display the plot
plt.show()

```

The code creates a bar chart to visualize the gender distribution in the dataset, labeling the x-axis as "Female" and "Male" and displaying the count of each category on top of the bars.

```python
# Set the figure size to 8x6 inches for better visibility
plt.figure(figsize=(8, 6))

# Create a histogram for the 'trtbps' (resting blood pressure) column
# Include a Kernel Density Estimate (KDE) line and set color to sky blue
sns.histplot(data['trtbps'], kde=True, bins=20, color='skyblue')

# Set the title and axis labels of the plot
plt.title('Distribution of Resting Blood Pressure')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

Resting Blood Pressure (trtbps): This variable represents the systolic blood pressure (in mm Hg) of the patient while at rest. It is an important health measure as it can help indicate the risk of cardiovascular disease and other health conditions.

```python
# Set the figure size to 8x6 inches
plt.figure(figsize=(8, 6))

# Create a count plot of the 'fbs' (fasting blood sugar) column using a pastel color palette
sns.countplot(x='fbs', data=data, palette='pastel')

# Replace default tick labels (0 and 1) with descriptive labels
plt.xticks([0, 1], ['Normal', 'High (> 120 mg/dL)'])

# Set the title and axis labels
plt.title('Distribution of Fasting Blood Sugar')
plt.xlabel('Fasting Blood Sugar')
plt.ylabel('Count')

# Annotate each bar with its count value
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}',           # Text: the height (count) of the bar
                       (p.get_x() + p.get_width() / 2,     # X-coordinate: center of the bar
                        p.get_height() + 1),               # Y-coordinate: just above the bar
                       ha='center', va='bottom')           # Center the text horizontally, align to bottom vertically

# Display the plot
plt.show()
```

This variable indicates whether the patient's fasting blood sugar level is greater than 120 mg/dL. Fasting blood sugar levels higher than 120 mg/dL are considered a risk factor for diabetes.
0 means the fasting blood sugar is normal (≤ 120 mg/dL).
1 means the fasting blood sugar is high (> 120 mg/dL).

```python
# Set the figure size to 6x4 inches
plt.figure(figsize=(6, 4))

# Create a boxplot of the 'chol' (serum cholesterol) column with a custom color
sns.boxplot(x=data['chol'], color='lightseagreen')

# Set the title and x-axis label of the plot
plt.title('Boxplot of Serum Cholesterol')
plt.xlabel('Cholesterol (mg/dL)')

# Adjust layout to prevent clipping of labels/titles
plt.tight_layout()

# Display the plot
plt.show()
```

A boxplot (or box-and-whisker plot) succinctly summarizes a continuous variable’s distribution by showing:
*Median (the 50th percentile) as the line inside the box
*Interquartile Range (IQR)—the box from the 25th to the 75th percentile
*Whiskers extending to the furthest points within 1.5 × IQR of the box
*Outliers plotted individually beyond the whiskers

```python
# Set the figure size to 6x4 inches
plt.figure(figsize=(6, 4))

# Create a count plot of the 'restecg' (resting ECG) column with pastel colors
ax = sns.countplot(x='restecg', data=data, palette='pastel')

# Set custom labels for the x-axis categories (0, 1, 2)
ax.set_xticklabels(['Normal (0)', 'ST-T Abnormality (1)', '         Left Ventricular Hypertrophy'])

# Set the title and axis labels
plt.title('Resting ECG Results Distribution')
plt.xlabel('Resting ECG Category')
plt.ylabel('Count')

# Annotate each bar with its count value
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',                   # Text: the bar height (count)
                (p.get_x() + p.get_width() / 2,             # X-coordinate: center of bar
                 p.get_height() + 1),                       # Y-coordinate: slightly above bar
                ha='center', va='bottom', fontsize=11)      # Align text to center and bottom

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

The restecg column represents the patient’s resting electrocardiographic (ECG) results, coded as:

0 = Normal

1 = ST-T wave abnormality (e.g., showing possible ischemia)

2 = Left ventricular hypertrophy (by Estes’ criteria)



```python
# Get the count of each category in the 'slp' (slope) column
slp_counts = data['slp'].value_counts()

# Set the figure size to 5x5 inches
plt.figure(figsize=(5, 5))

# Create a pie chart showing the distribution of slope categories
# - Labels correspond to values 0, 1, 2 with descriptive names
# - autopct displays percentages with one decimal place
# - startangle=90 rotates the pie chart for better orientation
# - colors specify custom pastel shades
plt.pie(slp_counts, 
        labels=['Upsloping (0)', 'Flat (1)', 'Downsloping (2)'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['#66b3ff', '#99ff99', '#ffcc99'])

# Set the title of the plot
plt.title('Distribution of Slope of Peak Exercise ST Segment', fontsize=16)

# Display the pie chart
plt.show()
```

The slp column refers to the slope of the peak exercise ST segment, a feature in an electrocardiogram (ECG) taken during a stress test. The ST segment is the part of the ECG that represents the time between the end of the heart's contraction and the beginning of its relaxation. The slope of this segment can provide insights into the heart's response to exercise, which is useful for diagnosing heart disease.
Values:
0 = Upsloping: The ST segment rises above the baseline, which may indicate a normal heart response during exercise.
1 = Flat: The ST segment remains level, which can suggest possible ischemia (reduced blood flow) but is not as severe as downsloping.
2 = Downsloping: The ST segment slopes downward, which is often a strong indicator of ischemia or heart disease, especially when observed during
exercise.

```python
# Filter the dataset to include only rows where 'thall' is 1, 2, or 3
filtered_df = data[data['thall'].isin([1, 2, 3])]

# Get the count of each 'thall' category in the filtered data
thall_counts = filtered_df['thall'].value_counts()

# Set the figure size to 5x5 inches
plt.figure(figsize=(5, 5))

# Create a pie chart to show the distribution of thalassemia status (1, 2, 3)
# - Labels are the unique values in the 'thall' column
# - autopct shows percentage values on slices
# - startangle=90 rotates the pie for better orientation
# - colors assign custom shades to each slice
plt.pie(thall_counts, 
        labels=thall_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['#ff9999', '#66b3ff', '#99ff99'])

# Add a title to the pie chart
plt.title('Distribution of Thalassemia Status (Values 1, 2, 3)')

# Ensure the pie chart is perfectly circular
plt.axis('equal')

# Display the chart
plt.show()
```

The filters the dataset to include only rows where the 'thall' column has values 1, 2, or 3. It then creates a pie chart showing the distribution of these values, with percentage labels for each segment and a title indicating that the chart represents Thalassemia status.
